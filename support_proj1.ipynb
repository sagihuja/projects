```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://pubmed.ncbi.nlm.nih.gov/25118243/'
r = requests.get(url, allow_redirects=True)

open('art_cont', 'wb').write(r.content)

with open('art_cont') as f:
    html = f.read()

page = requests.get(url)
print(page.text)
soup = bs(page.content, "html.parser")
```

```python
soup = bs(html, 'html.parser') 
print(type(soup))

#print(soup.prettify())
```

```python
print(soup.title)
type(soup.title)

print(soup.title.string)
#print(soup.title.name)
```

```python
link = soup.a
print(link.prettify())

links = soup.find_all('a')
print(type(links))
len(links)

print(links[0].prettify())

for link in links:
    print(f"{link}: {link.string} - {link.get('href')}")
```
